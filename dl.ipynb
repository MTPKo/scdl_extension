## Use with scdl


```python
import scdl
import subprocess
from tqdm import tqdm
import time
```

```python
dllist=[
        'https://soundcloud.com/mpmcgn/sets/example',
]
```

```python
def get_files(url):
    result =subprocess.run(f'''scdl -l {url} --original-metadata --hide-progress --error''',shell=True, capture_output=True, text=True)
    return result
```

```python
idx=0
rate_err =0
#sleeptimer in s
timer= 10
with tqdm(total=len(dllist)) as pbar:
    pbar.set_postfix({'d/l ': dllist[idx]})
    pbar.update(0)
    while idx < len(dllist):
        url = dllist[idx]
        result = get_files(url)
        output = result.stderr
        if result.stderr != "":
            if "HTTP Error 429" in output:
                rate_err = rate_err + 1
                timer = timer * rate_err
                pbar.set_postfix({'Rate limit exceeded':f''' Waiting for { timer if timer < 61 else timer/60} {'seconds' if timer < 61 else 'minutes'}'''})
                pbar.update(0)
                time.sleep(timer)  # Wait for time
            else:
                pbar.set_postfix({"ERR ": result.stderr})
                break
        else:
            idx += 1
            timer= 10
            rate_err =0
            pbar.update(1)


```
